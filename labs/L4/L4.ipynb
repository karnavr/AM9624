Let's quickly recap some concepts:

#### Covariance Matrix

In essence, the covariance matrix is a square matrix that summarizes the variances and covariances between multiple variables within a dataset. If you have $n$ variables (features), the covariance matrix will be an $n \times n$ matrix.

The diagonal elements represent the variances of each variable, while the off-diagonal elements represent the covariances between pairs of variables. Note that the covariance matrix is real symmetric. The Spectral Theorem states that the eigenvectors of a real symmetric matrix are orthogonal. If you normalize these vectors and compose them into a matrix, the matrix has the properties that its inverse is also its transpose.

#### Change of basis matrix

Sometimes, we want to change the basis of a vector space. A common reason is to diagonalize the covariance matrix, which can be useful for dimensionality reduction or for understanding the principal components of the data. In general, for some vector $v$,

$$
\mathbf{v}^{\prime}=P^{-1} \mathbf{v}
$$

where $P$ is the change of basis matrix, whose columns are the new basis vectors expressed in terms of the old basis, and $P^{-1}$ is the inverse of the change of basis matrix.
